# Merge two dataframes

```python
df_country = pd.read_excel("Dataset\Country-Code.xlsx")
df_country.head()
```

```python
# left join
final_df = pd.merge(df, df_country, on="Country Code", how="left")
final_df.head()
```

```python
final_df.dtypes
```

```python
final_df.columns
```

# Plotting Pie Chart

```python
final_df.Country.value_counts()
```

```python
country_names = final_df.Country.value_counts().index
country_names
```

```python
country_val = country_names = final_df.Country.value_counts().values
country_val
```

```python
plt.pie(country_val[:3], labels=country_names[:3], autopct="%1.2f%%")
```

# Rating observations

```python
final_df.columns
```

```python
final_df.groupby(["Aggregate rating", "Rating color", "Rating text"]).size()
```

```python
ratings = final_df.groupby(["Aggregate rating", "Rating color", "Rating text"]).size().reset_index().rename(columns={0:"Rating Count"})
ratings
```

# Visualize rating data

```python
plt.rcParams["figure.figsize"] = (12,6)
sns.barplot(x="Aggregate rating", y="Rating Count", data=ratings)
```

```python
sns.barplot(x="Aggregate rating", y="Rating Count", hue="Rating color", palette=['white', 'red', 'orange', 'yellow', 'green', 'green'] ,data=ratings)
```

# Observation
1. max counts are non-rated
2. max ratings btw -> 2.5 - 3.4

```python
sns.countplot(x="Rating color", data=ratings, palette=['white', 'red', 'orange', 'yellow', 'green', 'green'])
```

# Observations
1. no.of records in orange is high
